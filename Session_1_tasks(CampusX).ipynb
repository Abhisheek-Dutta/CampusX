###Q4+:- You are given two coordinates (x1, y1) and (x2, y2) of a two-dimensional graph. Find the distance between them.

###Input : 
        x1, y1 = (3, 4) 
        x2, y2 = (4, 3)

###Output : 1.41421


###We will use the distance formula derived from Pythagorean theorem. The formula for distance between two point (x1, y1) and (x2, y2) is40

Distance = $\sqrt{(x2-x1)^{2} + (y2-y1)^{2}}$


image.png

````python
'Method 1: Without using the inbuilt library'

def distance(x1, y1, x2, y2):
   
  # Calculating distance
   
  return (((x2 - x1)**2 +(y2 - y1)**2)**0.5)
 
print( distance(3, 4, 4, 3))

#Time Complexity: O(1) -> Time taken by the above code
#Auxiliary Space: O(1) -> Space used by the above code

#Auxiliary Space is the extra space or temporary space used by an algorithm. 
#The Space Complexity of an algorithm is the total space taken by the algorithm 
#with respect to the input size. Space complexity includes both Auxiliary space 
#and space used by input.

#Time complexity is defined as the amount of time taken by an algorithm to run, 
#as a function of the length of the input. It measures the time taken to execute 
#each statement of code in an algorithm. It is not going to examine the total 
#execution time of an algorithm.
````

````python
'Method 2: Using the inbuilt library'

# distance between two points
 
import math
 
# Function to calculate distance
def distance(x1 , y1 , x2 , y2):
 
    # Calculating distance
    return math.sqrt(math.pow(x2 - x1, 2) +
                math.pow(y2 - y1, 2) * 1.0)
 
print("%.6f"%distance(3, 4, 4, 3))

#Time Complexity: O(1)
#Auxiliary Space: O(1)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

image.png


image.png

````python
n = int(input('Enter the number: '))

result = (n*(n+1)*(2*n+1))/6 #-> Using formula

print('The sum of squares of first n natural numbers: ', result)
````

````python
#Method 1: O(N) The idea is to run a loop from 1 to n and for each i, 1 <= i <= n, find i2 to sum.

# find sum of square of first n natural numbers
  
  
# Return the sum of square of first n natural numbers
def squaresum(n) :
  
    # Iterate i from 1 and n finding square of i and add to sum
    sm = 0
    for i in range(1, n+1) :
        sm = sm + (i * i)
      
    return sm

n = 5
print(squaresum(n))
````

````python
#Method 2: O(1)

# find sum of square of first n natural numbers
  
# Return the sum of square of first n natural numbers
def squaresum(n) :
    return (n * (n + 1) * (2 * n + 1)) // 6
  
n = 5
print(squaresum(n))
````
